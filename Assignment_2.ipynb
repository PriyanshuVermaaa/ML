```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.hist(df['City'],bins=20)
plt.xlabel('Population',fontsize=15)
plt.ylabel('Frequency',fontsize=15)
plt.title('Histogram for Population city Bases')
plt.show()
```

```python
X = df['Unemployment Rate'].values[:,None]
Y = df['Population (millions)'].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
#linear reggresion
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_test_predict = regressor.predict(x_test)
```

```python
y_train_predict = regressor.predict(x_train)
```

```python
plt.scatter(x_train,y_train, color="pink",label='Actual')
plt.plot(x_test, y_test_predict, color='blue',label='Pridicted')
plt.xlabel('Unemployment Rate')
plt.title('Unemployment Rate vs Population (millions)')
plt.ylabel('Population')
plt.legend()
plt.show()
```

```python
from sklearn import metrics
r2_score = metrics.r2_score(y_test,y_test_predict)
mse = metrics.mean_squared_error(y_test,y_test_predict)
mae = metrics.mean_absolute_error(y_test,y_test_predict)
print("R2 Score:",r2_score)
print("MSE:",mse)
print("MAE:",mae)
format(r2_score,'.4f')
format(mse,'.4f')
format(mae,'.4f')
```

```python
r2_score = metrics.r2_score(y_train,y_train_predict)
mse = metrics.mean_squared_error(y_train,y_train_predict)
mae = metrics.mean_absolute_error(y_train,y_train_predict)
print("R2 score: {0}\nMSE:{2}".format(r2_score,mse,mae))
```
