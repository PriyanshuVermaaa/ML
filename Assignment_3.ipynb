### Visualize the data using R/Python by plotting the graphs for assignment no. 1 and 2. Consider 
### suitable data set. Use Scatter plot, Bar plot, Box plot, Pie chart, Line Chart.

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
data = pd.read_csv('admission.csv')
```

```python
# Scatter Plot
plt.figure(figsize=(6,4))
plt.scatter(data['GRE Score'], data['CGPA'], c='orange', alpha=1,)
plt.xlabel('GRE Score')
plt.ylabel('CGPA')
plt.title('Scatter Plot')
```

```python
# Bar Plot
plt.figure(figsize=(6,4))
plt.bar(data['University Rating'], data['CGPA'], color= 'red')
plt.xlabel('University Rating')
plt.ylabel('CGPA')
plt.title('Bar Plot')
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(6,4))
sns.set_palette("pastel")
sns.boxplot(x=data['University Rating'], y=data['TOEFL Score'])
plt.title('Box Plot')
plt.show()
```

```python
sns.set_palette("pastel")
sns.boxplot(x = "GRE Score", data = data, width = 0.8, showmeans = True, color="orange" )
```

```python
# Pie Chart
research = data['Research'].value_counts()
labels=['Research (1)', 'No Research (0)']
explode=[0,0.05]
plt.pie(research, labels=labels, autopct='%1.1f%%',explode=explode, shadow=True, colors=['orange', 'skyblue',],)
plt.title('Pie Chart')
plt.show()
```

```python
# line chart
plt.plot(data["Serial No."], data["CGPA"])
plt.xlabel('Serial No.')
plt.ylabel('CGPA')
plt.title('Line Chart')
plt.show()
```
